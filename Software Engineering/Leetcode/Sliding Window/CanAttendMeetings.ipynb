# Can attend meetings
Given an array of meeting time interval objects consisting of start and end times `[[start_1,end_1],[start_2,end_2],...] (start_i < end_i)`, determine if a person could attend all meeting to their schedule without any conflicts.

# Example 1
```python
Input: intervals = [(0,30),(5,10),(15,20)]
Output: False
```

### Explanation
- (0,30) and (5,10) will conflict
- (0,30) and (15,20) will conflict

# Example 2
```python
Input: intervals = [(5,8),(9,15)]
Ouptut: True
```

### Note
(0,8),(8,10) is not considered a conflict at 8

# Constraints
- `0 <= intervals.length <= 500`
- `0 <= intervals[i].start < intervals[i].end <= 1,000,000`


````python
class Interval:
    def __init__(self, start, end):
        self.start = start
        self.end = end

class Solution:
    def canAttendMeetings(self, interv: list[Interval]) -> bool:
        def selectionSort(intervals):
            for ind in range(len(intervals)):
                min_index = ind
                for j in range(ind + 1, len(intervals)):
                    if intervals[j].start < intervals[min_index].start:
                        min_index = j
                (intervals[ind], intervals[min_index]) = (intervals[min_index], intervals[ind])
        
        intervs = selectionSort(interv)

        for i in range(len(intervs)):
            i1 = intervs[i-1]
            i2 = intervs[i]
            if i1.end > i2.start:
                return False
        return True
````

````python
#Test Cases
sol = Solution()
print(sol.canAttendMeetings([Interval(1, 2), Interval(3, 4)]))
assert not sol.canAttendMeetings([Interval(2, 5), Interval(3, 4)])
# [(0,30),(5,10),(15,20)]
assert not sol.canAttendMeetings([Interval(0,30), Interval(5,10), Interval(15,20)])
print("All Tests pass (Yes, no, no).")
````
