# Min Stack
Design a stack class that supports the `push`, `pop`, `top`, and `getMin` operations.
- `MinStack()` initializes the stack object
- `void push(int val)` pushes the element `val` onto the stack
- `void pop()` removes the element on the top of the stack
- `int top()` gets the top element of the stack
- `int getMin()` retrieves the minimum element in the stack

Each function should run in O(1) time.

## Example 1
```python
Input: ["MinStack", "push", 1, "push", 2, "push", 0, "getMin", "pop", "top", "getMin"]

Output: [null, null, null, null, 0, null, 2, 1]

Explanation:
minStack = MinStack()
minStack.push(1)
minStack.push(2)
minStack.push(0)
minStack.getMin() # return 0
minStack.pop()
minStack.top() # return 2
minStack.getMin() # return 1
```

````python
class MinStack:
    def __init__(self):
        self.stack = []
        self.length = 0

    def push(self, val:int) -> None:
        self.stack.append(val)
        self.length += 1

    def pop(self) -> None:
        if self.length == 0:
            return #Or this could raise an error
        # temp = self.stack[-1]
        del self.stack[-1]
        self.length -= 1

    def top(self) -> int:
        if self.length == 0:
            return -1
        return self.stack[-1]

    def getMin(self) -> int:
        lowest = float('inf')
        for i in range(self.length):
            lowest = min(self.stack[i], lowest)
        if lowest == float('inf'):
            return -1
        return lowest
````

````python
# Example 1
if __name__ == "__main__":
    minStack = MinStack()
    minStack.push(1)
    minStack.push(2)
    minStack.push(0)
    assert minStack.getMin() == 0
    print("minStack.getMin() == 0")
    minStack.pop()
    assert minStack.length == 2
    print("minStack.length == 2")
    assert minStack.top() == 2
    print("minStack.top() == 2")
    assert minStack.getMin() == 1
    print("minStack.getMin() == 1")
````
